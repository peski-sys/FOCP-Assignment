```python
# Beckett Pizza Plaza 4-for-3 Offer
# Introduction to Programming – Task 1

print("Beckett Pizza Plaza 4-for-3 Offer")
print("=" * 33)

prices = []
pizza_number = 1

while len(prices) < 4:
    try:
        value = input(f"\nEnter Price of Pizza #{pizza_number}: ")
        price = float(value)

        if price <= 0:
            print("Please enter a valid price!")
            continue

        prices.append(price)
        pizza_number += 1

    except ValueError:
        print("Please enter a valid price!")

# Calculate total and discount
original_total = sum(prices)
free_pizza = min(prices)
final_total = original_total - free_pizza

discount_percentage = (free_pizza / original_total) * 100

discount_percentage = round(discount_percentage)

print(f"\nOrder Total is £{final_total:.2f}, a fabulous discount of {discount_percentage}%!")
```
