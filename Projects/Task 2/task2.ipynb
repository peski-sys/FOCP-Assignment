```python
# Password Verification Program
# Introduction to Programming â€“ Task 2

import random

password = input("Enter your password: ")

# Check password length
if len(password) < 9:
    print("Password too short.")
    exit()

# Ask for three random character positions
for _ in range(3):
    position = random.randint(1, len(password))  # positions start at 1
    guess = input(f"Enter letter at position {position}: ")

    if guess == password[position - 1]:
        print("Correct")
    else:
        print("\nSecurity check failed.")
        exit()

print("Security check passed.")
```
