```python
capitals = {} #lists that will coinain the capitals 

while True: #using loop to ask the users to enter country and its capital
    country = input("Enter a country (or 'quit' to stop): ").strip().lower()

    if country == "quit":
        break

    if country in capitals:
        print(f"The capital of {country.title()} is {capitals[country]}")
    else:
        capital = input(f"I don't know the capital of {country.title()}. Please enter it: ")
        capitals[country] = capital
        print("Capital stored!")

print(capitals)
```
