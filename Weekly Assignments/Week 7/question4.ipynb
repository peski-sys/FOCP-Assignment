```python
def letter_frequency(message):
    message = message.lower()
    freq = {}

    for char in message:
        if char.isalpha():
            freq[char] = freq.get(char, 0) + 1

    sorted_letters = sorted(freq.items(), key=lambda x: x[1], reverse=True)

    for letter, count in sorted_letters[:6]:
        print(letter, ":", count)

text = "This is an example encrypted message"
letter_frequency(text)

```
