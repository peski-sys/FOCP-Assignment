```python
numberOfSweets = int(input("How Many Students?: "))
numberOfPupils = int(input("Required Group Size?: "))

eachStudentSweets = int(numberOfSweets / numberOfPupils)
leftOverSweets = numberOfSweets % numberOfPupils

print(f"Each student will receive {eachStudentSweets} sweets with {leftOverSweets} sweets left over.")
```
