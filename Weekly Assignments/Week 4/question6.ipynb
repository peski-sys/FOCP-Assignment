```python
def celsius_to_fahrenheit(c):
    return (c * 9 / 5) + 32


def fahrenheit_to_celsius(f):
    return (f - 32) * 5 / 9

temp = input("Enter temperature (e.g. 25C): ")

value = float(temp[:-1])
unit = temp[-1].upper()

if unit == "C":
    result = celsius_to_fahrenheit(value)
    print(f"{result:.1f}F")
else:
    print("Invalid input")

```
