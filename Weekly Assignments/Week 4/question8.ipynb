```python
from statistics import mean

temps = []

while True:
    temp = input("Enter temperature (or press Enter to finish): ")
    
    if temp == "":
        break

    value = float(temp[:-1])
    temps.append(value)

if temps:
    print("Maximum:", max(temps))
    print("Minimum:", min(temps))
    print("Mean:", mean(temps))
else:
    print("No temperatures entered")
```
