```python
def count_case(text):
    upper = 0
    lower = 0

    for char in text:
        if char.isupper():
            upper += 1
        elif char.islower():
            lower += 1

    return upper, lower


sample = input("Enter a string: ")
upper_count, lower_count = count_case(sample)

print("Uppercase letters:", upper_count)
print("Lowercase letters:", lower_count)
```
