```python
def is_prime(n):#checks if the given number is prime or not
    if n <= 1: # 0 or negative
        return False

    for i in range(2, n): #checks if the factors are more than 2
        if n % i == 0:
            return False

    return True


print(is_prime(18))  
print(is_prime(11))  
```
