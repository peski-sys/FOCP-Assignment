```python
import random
def hide_message(message):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    message = message.replace(" ", "")

    interval = random.randint(2, 20)
    encrypted = ""

    for char in message:
        encrypted += char
        for _ in range(interval - 1):
            encrypted += random.choice(alphabet)

    return encrypted, interval


def decrypt_message(encrypted, interval):
    decrypted = ""

    for i in range(0, len(encrypted), interval):
        decrypted += encrypted[i]

    return decrypted



encrypted, interval = hide_message("send cheese")
print("Encrypted:", encrypted)
print("Decrypted:", decrypt_message(encrypted, interval))
```
