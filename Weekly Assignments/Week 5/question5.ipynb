```python
import sys
from statistics import mean

args = sys.argv[1:]

if not args:
    print("Error: No temperatures provided")
    sys.exit()

temps = []

for temp in args:
    try:
        value = float(temp[:-1])
        temps.append(value)
    except ValueError:
        print(f"Invalid temperature format: {temp}")
        sys.exit()

print("Maximum:", max(temps))
print("Minimum:", min(temps))
print("Mean:", mean(temps))
```
